# 取得したデータ一覧

- 加速度（xyzの三軸）
- 角速度（xyzの三軸）
- 移動量（xyの二軸）

合計 8種類の特徴

```python
import pandas as pd
from datetime import datetime
```

```python
def convert_timestamp(ts):
    ts = ts / 1000
    return datetime.fromtimestamp(ts)
```

```python
df_accel = pd.read_csv('./csv/exabe2LogASma.csv', header=None, index_col=3, names=['x_accel', 'y_accel', 'z_accel'])
df_accel.index.name = 'time'
df_gyro = pd.read_csv('./csv/exabe2LogGSma.csv', index_col='time')
df_inc = pd.read_csv('./csv/exabe2LogISma.csv', index_col='time')
```

```python
df_accel.index = df_accel.index.map(convert_timestamp)
df_gyro.index = df_gyro.index.map(convert_timestamp)
df_inc.index = df_inc.index.map(convert_timestamp)
```

### 加速度のデータ

### 角速度のデータ

```python
df_gyro
```

### 移動量のデータ
